```python
import os
import requests
from dotenv import load_dotenv

load_dotenv()

HF_API_KEY = os.getenv("HUGGINGFACE_API_KEY")

API_URL = "https://router.huggingface.co/v1/chat/completions"

headers = {
            "Authorization": f"Bearer {HF_API_KEY}",
            "Content-Type": "application/json",
        }
```

```python
def ask_llm(prompt):
    payload = {
        "model": "deepseek-ai/DeepSeek-V3",
        "messages": [
            {
                "role": "user",
                "content": prompt
            }
        ],
        "temperature": 0.7,
        "max_tokens": 512
    }

    response = requests.post(API_URL, headers=headers, json=payload)

    if response.status_code != 200:
        raise Exception(response.text)

    return response.json()["choices"][0]["message"]["content"]
```

## Experiment 1 – Role Prompting

```python
prompt = """
You are an experienced Artificial Intelligence professor.

Explain Machine Learning to a beginner using simple language and real-world examples.
"""

print("========== Experiment ==========\n")
print(prompt)

print("\n========== OUTPUT ==========\n")
print(ask_llm(prompt))
```

### Observations
- The model adopted the role of an AI professor and responded accordingly.
- he explanation was simple, organized, and easy for beginners to understand.
- Real-world examples made the concept more relatable.
- Assigning a role improved the tone and quality of the response.
### Improvements
- Specify the target audience more clearly (e.g., school student or university student).
- Set a word limit to keep the response concise.
- Request examples or analogies for better understanding.
- Specify the desired output format, such as bullet points.

## Experiment 2 – Chain-of-Thought Prompting

```python
prompt = """
A store sold 250 smartphones.

40% were Android phones.

The remaining phones were iPhones.

Think step by step and calculate how many iPhones were sold.
"""

print("========== Experiment ==========\n")
print(prompt)

print("\n========== OUTPUT ==========\n")
print(ask_llm(prompt))
```

### Observations
- The model solved the problem step by step.
- Each calculation was clearly explained before the final answer.
- The reasoning process made the solution easy to verify.
- The structured reasoning reduced the chance of errors.
### Improvements
- Ask the model to label each reasoning step.
- Request the final answer to be highlighted separately.
- Specify that all calculations should be shown.
- Include units where applicable for clarity.

## Experiment 3 – Few-Shot Prompting

```python
prompt = """
Example 1

Input: Python
Output: Programming Language

Example 2

Input: Tesla
Output: Company

Example 3

Input: Lion
Output: Animal

Now classify:

Input: Google
"""

print("========== Experiment ==========\n")
print(prompt)

print("\n========== OUTPUT ==========\n")
print(ask_llm(prompt))
```

### Observations
- The provided examples helped the model recognize the expected pattern.
- The classification was accurate and consistent.
- The output followed the same format as the examples.
- Few-shot prompting improved response consistency.
### Improvements
- Add more examples covering different categories.
- Include examples with similar inputs to reduce ambiguity.
- Clearly define the expected output format.
- Use diverse examples to improve generalization.

## Experiment 4 – Structured Output (JSON)

```python
prompt = """
Provide information about Pakistan.

Return ONLY valid JSON.

Include:
- country
- capital
- currency
- official_language
- population
"""

print("========== Experiment ==========\n")
print(prompt)

print("\n========== OUTPUT ==========\n")
print(ask_llm(prompt))
```

### Observations
- The response was returned in valid JSON format.
- The output was easy to read and machine-process.
- Only the requested fields were included.
- Structured output is useful for APIs and software applications.
### Improvements
- Specify the required data types for each field.
- Request valid JSON without Markdown formatting.
- Define whether values should be strings, integers, or arrays.
- Include formatting rules to ensure consistency.

## Experiment 5 – Prompt Optimization

 -- Before Optimization

```python
prompt = "Tell me about Artificial Intelligence."

print("========== Experiment ==========\n")
print(prompt)

print("\n========== OUTPUT ==========\n")
print(ask_llm(prompt))
```

-- After Optimization

```python
prompt = """
Explain Artificial Intelligence to a first-year Computer Science student.

Limit your response to 200 words.

Include:
- Definition
- Three real-world applications
- Advantages
- Challenges

Format your answer using bullet points.
"""

print("========== Experiment ==========\n")
print(prompt)

print("\n========== OUTPUT ==========\n")
print(ask_llm(prompt))
```

### Observations
- The original prompt produced a broad and generic response.
- The optimized prompt generated a more focused and organized answer.
- Adding constraints improved response quality.
- The optimized response better matched the intended audience and purpose.
### Improvements
- Clearly define the target audience.
- Specify the desired response length.
- Mention the required topics to cover.
- Request a specific format such as bullet points or numbered lists.
- Remove ambiguity by providing detailed instructions.
